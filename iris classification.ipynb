# IRIS FLOWER CLASSIFIER

The Iris flower dataset is a famous set of data that contains multiple properties of Iris flowers. The IRIS flowers are classified using their attributes like sepal length, sepal width, petal length, and petal width. To make these classifications accurate, i have trained a machine learning model that classifies new iris flowers based on thier properties.

```python
import warnings
warnings.filterwarnings('ignore')
```

# IMPORTING MODULES

```python
import pandas as pd
#import matplotlib.pyplot as plt
#%matplotlib inline
import seaborn as sns
sns.set(style="white", color_codes=True)
```

# IMPORTING DATASET

```python
iris= pd.read_csv(".\Downloads\Iris.csv")
iris.head()
```

# Dropping Unwanted Columns

```python
iris=iris.drop(columns=['Id'])
iris.head()
```

```python
iris["Species"].value_counts()
```

# EXPLORATORY DATA ANALYSIS USING SEABORN

```python
sns.pairplot(iris,hue="Species")
```

```python
sns.FacetGrid(iris,hue="Species",height=6).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend()
```

# CORRELATION MATRIX

```python
iris.corr()
```

```python
corr = iris.corr()
sns.heatmap(corr,cmap='coolwarm')
```

# Converting Categorical Value into Numeric Form

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
```

```python
iris['Species']=l.fit_transform(iris['Species'])
iris.head()
```

```python
sns.heatmap(iris)
```

# TEST AND TRAIN DATA SPLITTING

```python
from sklearn.model_selection import train_test_split
x=iris.drop(columns=['Species'])
y=iris['Species']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print("Accuracy:",model.score(x_test,y_test)*100)
```

# KNeighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
print("Accuracy:",model2.score(x_test,y_test)*100)
```

# Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier()
```

```python
model3.fit(x_train,y_train)
```

```python
print("Accuracy:",model3.score(x_test,y_test)*100)
```
